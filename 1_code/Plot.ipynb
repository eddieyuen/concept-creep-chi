# relativefreq_concepts

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Apply moving smoothing average of 5 years
def moving_average(data, window=5):
    return data.rolling(window=window, center=True, min_periods=1).mean()

# Specify file path
relative_frequency_file_path = '/Users/kawaiyuen/nlpworkshop/concept-creep-chi/2_pipeline/tmp/relativefreq_concepts.csv'

# Read relative frequency data
relative_frequency_data = pd.read_csv(relative_frequency_file_path, index_col=0)

# Step 1: Apply moving smoothing average of 3 years
smoothed_data = moving_average(relative_frequency_data)

# Plot line graph
plt.figure(figsize=(10, 6))  # Adjust the figure size as per your preference
for column in smoothed_data.columns:
    plt.plot(smoothed_data.index, smoothed_data[column], label=column)

plt.xlabel('Year')
plt.ylabel('Relative Frequency')
plt.title('Frequency of Target Concepts Over Time (Smoothed)')
plt.legend()
plt.grid(True)
plt.show()
```
